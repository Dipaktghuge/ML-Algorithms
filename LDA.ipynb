```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
data_wine=load_wine()
```

```python
ind=data_wine.data
tar=data_wine.target
```

```python
np.unique(tar)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA= LinearDiscriminantAnalysis()
```

```python
LDA_transformed=LDA.fit_transform(ind,tar)
```

```python
LDA.explained_variance_ratio_
```

```python
import matplotlib.pyplot as myplt
myplt.xlabel('LDA1')
myplt.ylabel('LDA2')
myplt.scatter(LDA_transformed[:,0],LDA_transformed[:,1],c=tar,cmap='rainbow',edgecolors='g')
```
